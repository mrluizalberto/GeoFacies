```python
# Define a color table
colorTable = ((0,0,143,255), (0,0,159,255), (0,0,175,255), (0,0,191,255), (0,0,207,255),
              (0,0,223,255), (0,0,239,255), (0,0,255,255), (0,16,255,255), (0,32,255,255),
              (0,48,255,255), (0,64,255,255), (0,80,255,255), (0,96,255,255), (0,112,255,255),
              (0,128,255,255), (0,143,255,255), (0,159,255,255), (0,175,255,255), (0,191,255,255),
              (0,207,255,255), (0,223,255,255), (0,239,255,255), (0,255,255,255), (16,255,255,255),
              (32,255,239,255), (48,255,223,255), (64,255,207,255), (80,255,191,255), (96,255,175,255),
              (112,255,159,255), (128,255,143,255), (143,255,128,255), (159,255,112,255), (175,255,96,255),
              (191,255,80,255), (207,255,64,255), (223,255,48,255), (239,255,32,255), (255,255,16,255),
              (255,255,0,255), (255,239,0,255), (255,223,0,255), (255,207,0,255), (255,191,0,255),
              (255,175,0,255), (255,159,0,255), (255,143,0,255), (255,128,0,255), (255,112,0,255),
              (255,96,0,255), (255,80,0,255), (255,64,0,255), (255,48,0,255), (255,32,0,255),
              (255,16,0,255), (255,0,0,255), (239,0,0,255), (223,0,0,255), (207,0,0,255),
              (191,0,0,255), (175,0,0,255), (159,0,0,255), (143,0,0,255))
    
# Plot an image
def plot_image(mat, colorTable=colorTable):


    
    # Get image dimensions
    (x,y) = mat.shape
    
    # Get the maximum and minimum values
    minValue = np.amin(mat)
    maxValue = np.amax(mat)

    # Define min and max color indexes
    minIndex = 0
    maxIndex = len(colorTable)-1
    
    # Generate the color index volume
    img_color = np.zeros((x,y), dtype=int)
    img_color = np.round((mat-minValue)/(maxValue-minValue)*(maxIndex-minIndex)+minIndex).astype(int)
        
    # Create and display image
    img = Image.new( 'RGBA', (x,y), "white") # create a new white image
    pixels = img.load() # create the pixel map

    for i in range(img.size[0]):  # for every pixel
        for j in range(img.size[1]):
            pixels[i,j] = (colorTable[img_color[i,j]][0], 
                           colorTable[img_color[i,j]][1], 
                           colorTable[img_color[i,j]][2], 
                           colorTable[img_color[i,j]][3])
    imshow(np.array(img), origin='lower')
```

```python
def dist(x, y):
    return math.sqrt(np.dot(x, x) - 2 * np.dot(x, y) + np.dot(y, y))
```

```python
# Load ip, porosity, and facies images
ip = Vol()
phi = Vol()
ip.read("../GeoFacies-Data/Real_data/real_impedance_slice.vol")
phi.read("../GeoFacies-Data/Real_data/real_porosity_slice.vol")
```

```python
# Change image size
sizeX = ip.sizeX
sizeY = ip.sizeY
```

```python
# Create ip, facies, and phi image matrices
ip_mat = ip.data.reshape(ip.sizeY,ip.sizeX).T
ip_mat = ip_mat[:sizeX,:sizeY]
phi_mat = phi.data.reshape(phi.sizeY,phi.sizeX).T
phi_mat = phi_mat[:sizeX,:sizeY]
```

```python
# Create network
G = nx.Graph()
R = 4  # radius distance to neighbors


# Add the node values and edges in radius R
node_list = []
for x in range(sizeX):
    for y in range(sizeY):
        G.add_node((x,y))
        node_list.append((x,y))
        u = [x,y]
        for i in range(x,x+R+1):
            for j in range(y-R,y+R+1):
                if (i >= 0 and i < sizeX) and (j >= 0 and j < sizeY):
                    v = [i,j]
                    d = euclidean_distances(u,v)
                    if d > 0. and d <= R:
                        G.add_edge((x,y),(i,j))
```

```python
# Obtain the adjacency matrix representation
A = np.array(nx.to_numpy_matrix(G,nodelist=node_list))
#A = np.array((nx.adjacency_matrix(G,nodelist=node_list)).todense())
```

```python
# Run experiment REPS times
REPS = 1
train_frac = 0.3 # training fraction over the whole dataset
errs = np.zeros((REPS,6,6))

# Create arrays to keep results
flx_facies_mat = np.zeros((REPS, sizeX,sizeY))
tcrfri_facies_mat = np.zeros((REPS, sizeX,sizeY))
flx_phi_mat = np.zeros((REPS, sizeX,sizeY))
svr_phi_mat = np.zeros((REPS, sizeX,sizeY))
rr_phi_mat = np.zeros((REPS, sizeX,sizeY))
tr_phi_mat = np.zeros((REPS, sizeX,sizeY))
tcrfri_phi_mat = np.zeros((REPS, sizeX,sizeY))
test_size = np.floor(sizeY*sizeX*(1.-train_frac))
pred_flx = np.zeros((REPS,test_size))
pred_svr = np.zeros((REPS,test_size))
pred_rr = np.zeros((REPS,test_size))
pred_tr = np.zeros((REPS,test_size))
pred_tcrfri = np.zeros((REPS,test_size))
truth = np.zeros((REPS,test_size))
X_inputs = np.zeros((REPS,test_size))
        
for r in range(REPS):
    # Build X and y arrays
    X = []
    y = []
    facies = []
    img_coord = np.zeros((sizeX*sizeY,2),dtype=int)
    k = 0
    for i in range(sizeX): 
        for j in range(sizeY):
            X.append([ip.data[0,j,i]])    
            y.append(phi.data[0,j,i])
            img_coord[k][0] = i
            img_coord[k][1] = j
            k+=1
            
    X = np.array(X, copy=True)
    print X.shape
    X = X[:,0]
    X = X.reshape(X.size, 1)
    print X.shape
    y = np.concatenate(np.array(y, copy=True).reshape((-1,1)))
    print y.shape
    facies = np.array(facies)  
    
    # Define the training and testing sets
    samples = X.shape[0]
    inds = np.random.permutation(range(samples))
    train = inds[:np.floor(samples*train_frac)]
    test = inds[np.floor(samples*train_frac):]
        
    # Normalize the data (zero mean and unit variance)
    X = X - np.mean(X[train,:])
    y = y - np.mean(y[train])
    X /= np.max(np.abs(X[train,:]))
    y /= np.max(np.abs(y[train]))
    y *= 10.
    truth[r, :] = np.array(y[test], copy=True)
    X_inputs[r, :] = np.array(X[test, 0], copy=True)

    # Scale the data
    # X[train]*=1.
    # X[test]*=1.
    #y[train]*=1.
    #y[test]*=1.
    y_truth = np.array(y[test], copy=True)
        
    # Add the intercept to X
    X = np.hstack((X, np.ones((X.shape[0],1))))
    print X.shape
    
    # Define some auxiliary variables
    feats = X.shape[1] # number of features
    states = 3 # number of facies
    names = []
    
    #plt.figure(1)
    #plt.plot(X[:, 0], X[:,1], '.b')
    #plt.show()
    
    print np.max(np.abs(y))
    # Run the methods
    print 'Iteration '+ str(r)
    (name, pred_flx[r, :], lats_flx) = method_flexmix(np.array(X, copy=True), np.array(y, copy=True), train, test, states=states)
    (name, pred_svr[r, :], lats_svr) = method_svr(np.array(X, copy=True), np.array(y, copy=True), train, test, states=states)
    (name, pred_rr[r, :], lats_rr) = method_ridge_regression(np.array(X, copy=True), np.array(y, copy=True), train, test, states=states, params=0.00001)
    (name, pred_tr[r, :], lats_tr) = method_transductive_regression(np.array(X, copy=True), np.array(y, copy=True), train, test, states=states, params=[0.00001, 0.5, 0.5])
    (name, pred_tcrfri[r, :], lats_tcrfri) = method_tcrfr_indep(np.array(X, copy=True), np.array(y, copy=True), train, test, A, states=states, params=[0.99, 0.00001, 0.2], plot=True)
    
    # Evaluate the performance
    foo = [0,0]
    (errs[r,0, :],_) = evaluate(truth[r, :], pred_flx[r, :], foo, foo)
    (errs[r,1, :],_) = evaluate(truth[r, :], pred_svr[r, :], foo, foo)
    (errs[r,2, :],_) = evaluate(truth[r, :], pred_rr[r, :], foo, foo)
    (errs[r,3, :],_) = evaluate(truth[r, :], pred_tr[r, :], foo, foo)
    (errs[r,4, :],_) = evaluate(truth[r, :], pred_tcrfri[r, :], foo, foo) 
    
    # Store image results
    for rk in range(len(train)):
        ri = img_coord[train[rk]][0]
        rj = img_coord[train[rk]][1]
        flx_phi_mat[r,ri,rj] = y[train[rk]]
        svr_phi_mat[r,ri,rj] = y[train[rk]]
        rr_phi_mat[r,ri,rj] = y[train[rk]]
        tr_phi_mat[r,ri,rj] = y[train[rk]]
        tcrfri_phi_mat[r,ri,rj] = y[train[rk]]
    for rk in range(len(test)):
        ri = img_coord[test[rk]][0]
        rj = img_coord[test[rk]][1]
        flx_facies_mat[r,ri,rj] = lats_flx[rk]
        tcrfri_facies_mat[r,ri,rj] = lats_tcrfri[rk]
        flx_phi_mat[r,ri,rj] = pred_flx[r,rk]
        svr_phi_mat[r,ri,rj] = pred_svr[r,rk]
        rr_phi_mat[r,ri,rj] = pred_rr[r,rk]
        tr_phi_mat[r,ri,rj] = pred_tr[r,rk]
        tcrfri_phi_mat[r,ri,rj] = pred_tcrfri[r,rk]
        
```

```python
# Get averages
err_avg = np.mean(errs,axis=0)
flx_facies_mat_avg = np.mean(flx_facies_mat,axis=0)
tcrfri_facies_mat_avg = np.mean(tcrfri_facies_mat,axis=0)
flx_phi_mat_avg = np.mean(flx_phi_mat,axis=0)
svr_phi_mat_avg = np.mean(svr_phi_mat,axis=0)
rr_phi_mat_avg = np.mean(rr_phi_mat,axis=0)
tr_phi_mat_avg = np.mean(tr_phi_mat,axis=0)
tcrfri_phi_mat_avg = np.mean(tcrfri_phi_mat,axis=0)
```
