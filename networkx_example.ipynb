```python
import os
import sys
import numpy as np
import random
import matplotlib.pyplot as plt
import networkx as nx

from volume import Vol
```

```python
%matplotlib inline
```

```python
# Load ip and porosity volumes
ip = Vol()
phi = Vol()
ip.read("../GeoFacies-Data/ip_112_150x200x1.vol")
phi.read("../GeoFacies-Data/phi_112_150x200x1.vol")
```

```python
# Plot the porosity image results
f = plt.figure(figsize=(9, 7))
ax = f.add_subplot(1, 2, 1)
ip.plotSlice(0)
ax.set_title('Impedance')
ax = f.add_subplot(1, 2, 2)
phi.plotSlice(0)
ax.set_title('Porosity')
plt.show()
```

```python
G = nx.Graph()
```

```python
# Add the node values and edges
# Remember that 'weight' is a reserved word and used by many algorithms in NetworkX
z = 0
for x in range(ip.sizeX):
    for y in range(ip.sizeY):
        G.add_node((x,y,z), weight = random.random(), ip=ip.data[z,y,x], phi=phi.data[z,y,x])
        if x < ip.sizeX-1:
            G.add_edge((x,y,z),(x+1,y,z))
        if y < ip.sizeY-1:    
            G.add_edge((x,y,z),(x,y+1,z))
```

```python
# Attributes in nodes
G.node[(90,0,0)]['ip'],G.node[(0,0,0)]['phi']
```

```python
# Adjacency representation of graph as a dictionary of lists
a = nx.to_dict_of_lists(G)
```
