```python
def calc_error(truth, preds):
    # return root_mean_squared_error(truth, preds)
    return mean_absolute_error(truth, preds)
    # return r2_score(truth, preds)
```

```python
# Plot an image
def plot_image(mat):

    # Define a color table
    colorTable = (
              (0,0,143,255), (0,0,159,255), (0,0,175,255), (0,0,191,255), (0,0,207,255),
              (0,0,223,255), (0,0,239,255), (0,0,255,255), (0,16,255,255), (0,32,255,255),
              (0,48,255,255), (0,64,255,255), (0,80,255,255), (0,96,255,255), (0,112,255,255),
              (0,128,255,255), (0,143,255,255), (0,159,255,255), (0,175,255,255), (0,191,255,255),
              (0,207,255,255), (0,223,255,255), (0,239,255,255), (0,255,255,255), (16,255,255,255),
              (32,255,239,255), (48,255,223,255), (64,255,207,255), (80,255,191,255), (96,255,175,255),
              (112,255,159,255), (128,255,143,255), (143,255,128,255), (159,255,112,255), (175,255,96,255),
              (191,255,80,255), (207,255,64,255), (223,255,48,255), (239,255,32,255), (255,255,16,255),
              (255,255,0,255), (255,239,0,255), (255,223,0,255), (255,207,0,255), (255,191,0,255),
              (255,175,0,255), (255,159,0,255), (255,143,0,255), (255,128,0,255), (255,112,0,255),
              (255,96,0,255), (255,80,0,255), (255,64,0,255), (255,48,0,255), (255,32,0,255),
              (255,16,0,255), (255,0,0,255), (239,0,0,255), (223,0,0,255), (207,0,0,255),
              (191,0,0,255), (175,0,0,255), (159,0,0,255), (143,0,0,255))
    
    # Get image dimensions
    (x,y) = mat.shape
    
    # Get the maximum and minimum values
    minValue = np.amin(mat)
    maxValue = np.amax(mat)

    # Define min and max color indexes
    minIndex = 0
    maxIndex = len(colorTable)-1
    
    # Generate the color index volume
    img_color = np.zeros((x,y), dtype=int)
    img_color = np.round((mat-minValue)/(maxValue-minValue)*(maxIndex-minIndex)+minIndex).astype(int)
        
    # Create and display image
    img = Image.new( 'RGBA', (x,y), "white") # create a new white image
    pixels = img.load() # create the pixel map

    for i in range(img.size[0]):  # for every pixel
        for j in range(img.size[1]):
            pixels[i,j] = (colorTable[img_color[i,j]][0], 
                           colorTable[img_color[i,j]][1], 
                           colorTable[img_color[i,j]][2], 
                           colorTable[img_color[i,j]][3])
    imshow(np.array(img), origin='lower')
```

```python
# Load ip, porosity, and facies images
ip = Vol()
ei = Vol()
facies = Vol()
llambda = Vol()
lmu = Vol()
pr = Vol()
phi = Vol()
si = Vol()
ip.read("../GeoFacies-Data/ip_112_150x200x1.vol")
ei.read("../GeoFacies-Data/ei_112_150x200x1.vol")
facies.read("../GeoFacies-Data/facies_112_150x200x1.vol")
llambda.read("../GeoFacies-Data/llambda_112_150x200x1.vol")
lmu.read("../GeoFacies-Data/lmu_112_150x200x1.vol")
pr.read("../GeoFacies-Data/pr_112_150x200x1.vol")
phi.read("../GeoFacies-Data/phi_112_150x200x1.vol")
si.read("../GeoFacies-Data/si_112_150x200x1.vol")
#phi.read("phi_8_100x130x1.vol")
#ip.read("ip_8_100x130x1.vol")
#facies.read("facies_8_100x130x1.vol")
```

```python
# Create ip and facies image matrices
ip_mat = ip.data.reshape(ip.sizeY,ip.sizeX).T
facies_mat = facies.data.reshape(facies.sizeY,facies.sizeX).T
```

```python
# Get top and base Z locations of the reservoir
resTop = ip.getResTop()
resBase = ip.getResBase()
```

```python
# Set the training and testing coordinate sets
train_samples_perc = 0.15  # training samples percentage over the whole dataset
locXY_train = []
locXY_test = []

for i in range(ip.sizeX): 
    for j in range(ip.sizeY):
        locXY_test.append((i,j))

n = int(len(locXY_test)*train_samples_perc) # Number of training samples
for i in range(n):
    n_elm = len(locXY_test)
    elm = random.randint(0,n_elm-1) # Randomly choose the training pixel coordinates
    x = locXY_test[elm][0]
    y = locXY_test[elm][1]
    locXY_train.append((x,y))
    locXY_test.remove((x,y))
```

```python
# Build the training set
X_train_list = []
y_train_list = []
facies_train_list = []

n = len(locXY_train)
for k in range(n):
    x = locXY_train[k][0]
    y = locXY_train[k][1]
    for i in range(ip.sizeZ):
        if i >= resTop[x,y] and i <= resBase[x,y]:
            X_train_list.append([ip.data[i,y,x],ei.data[i,y,x],llambda.data[i,y,x],
                                 lmu.data[i,y,x],pr.data[i,y,x],si.data[i,y,x]])
            y_train_list.append(phi.data[i,y,x])
            facies_train_list.append(int(facies.data[i,y,x]))
        
X_train = np.array(X_train_list)
scalerX = preprocessing.StandardScaler().fit(X_train)
X_train = scalerX.transform(X_train)
y_train = np.concatenate(np.array(y_train_list).reshape((-1,1)))
scalerY = preprocessing.StandardScaler().fit(y_train)
y_train = scalerY.transform(y_train)
facies_train = np.array(facies_train_list)

# Add the intercept to X
aux = np.ones((X_train.shape[0],1))
X_train = np.hstack((X_train,aux))
print X_train.shape
```

```python
# Build the testing set
X_test_list = []
y_test_list = []
facies_test_list = []
n = len(locXY_test)
for k in range(n):
    x = locXY_test[k][0]
    y = locXY_test[k][1]
    for i in range(ip.sizeZ):
        if i >= resTop[x,y] and i <= resBase[x,y]:          
            X_test_list.append([ip.data[i,y,x],ei.data[i,y,x],llambda.data[i,y,x],
                                lmu.data[i,y,x],pr.data[i,y,x],si.data[i,y,x]])
            y_test_list.append(phi.data[i,y,x])
            facies_test_list.append(int(facies.data[i,y,x]))
    
X_test = np.array(X_test_list)
X_test = scalerX.transform(X_test)
y_test = np.concatenate(np.array(y_test_list).reshape((-1,1)))
y_test = scalerY.transform(y_test)
facies_test = np.array(facies_test_list)

# Add the intercept to X
aux = np.ones((X_test.shape[0], 1))
X_test = np.hstack((X_test,aux))
print X_test.shape
```

```python
# Define some variables
feats = X_train.shape[1] # number of features
states = 4 # number of facies
```

```python
# Train ordinary ridge regression
E = np.zeros((feats, feats))
np.fill_diagonal(E, 0.00001)
XXt = X_train.T.dot(X_train) + E
XtY = (X_train.T.dot(y_train))
w = np.linalg.inv(XXt).dot(XtY)
y_pred_rr = w.T.dot(X_test.T).T
rr_mse = calc_error(y_test, y_pred_rr)
```

```python
# Train ordinary support vector regression
clf = SVR(C=1.0, epsilon=0.1, kernel='linear', shrinking=False)
clf.fit(X_train, y_train.reshape(-1,))
y_pred_svr = clf.predict(X_test)
svr_mse = calc_error(y_test, y_pred_svr)
```

```python
# Set train and test data format for latent ridge and svr estimations
train_mc = MulticlassRegressionModel(co.matrix(X_train.T), classes=states, y=co.matrix(y_train))
test_mc = MulticlassRegressionModel(co.matrix(X_test.T), classes=states)
```

```python
# train latent support vector regression
lsvr = LatentRidgeRegression(theta=0.5, lam=0.0001, gam=1.)
(_, train_lats) = lsvr.fit(train_mc, max_iter=200)
lrr_fit_lbls = np.array(train_lats)
(y_pred_lrr, lats) = lsvr.predict(test_mc)
y_pred_lrr = np.array(y_pred_lrr)[:, 0]
lrr_pred_lbls = np.array(lats)
lrr_mse = calc_error(y_test, y_pred_lrr)
print 'LatentRR  mse: ', lrr_mse
```

```python
# Kmeans + Ridge Regression
kmeans = cl.KMeans(n_clusters=states, init='random', n_init=10, max_iter=100, tol=0.0001)
result = kmeans.fit(X_train)
kmeans_fit_lbls = result.labels_
sol = np.zeros((states, feats))
for i in range(states):
    inds = np.where(kmeans.labels_ == i)[0]
    ny = y_train[inds].reshape(len(inds), 1)
    nX = X_train[inds].reshape(len(inds), feats)
    foo = lsvr.train_model(nX, ny)
    sol[i, :] = np.array(foo).reshape(1, feats)

kmeans_pred_lbls = kmeans.predict(X_test)
y_pred_krr = np.sum(sol[kmeans_pred_lbls, :] * X_test, axis=1)
krr_mse = calc_error(y_test, y_pred_krr)
```

```python
# Use latent class regression FlexMix package from R
from rpy2.robjects.packages import importr
%load_ext rmagic
import rpy2.robjects as robjects
import rpy2.robjects.numpy2ri
import pandas.rpy.common as com
import pandas as pd
r = robjects.r
r.library("flexmix")

trainData = np.hstack((X_train[:,0:feats-1],y_train.reshape(-1,1)))
testData = np.hstack((X_test[:,0:feats-1],y_test.reshape(-1,1)))
df_train = pd.DataFrame(trainData)
df_test = pd.DataFrame(testData)

colNames = []
for i in range(feats-1):
    colNames.append(str(i))
colNames.append('y')
df_train.columns = colNames
df_test.columns = colNames

df_train_r = com.convert_to_r_dataframe(df_train)
df_test_r = com.convert_to_r_dataframe(df_test)

%R mycont <- list(iter = 200, tol = 0.001, class = "CEM")
%R as(mycont, "FLXcontrol")

model = r.flexmix(robjects.Formula("y ~ ."),data=df_train, k=states)

flexmix_fit_lbls = np.array(r.clusters(model))
flexmix_pred_lbls = np.array(r.clusters(model,newdata=df_test_r))

pr = r.predict(model, newdata=df_test_r)
df = com.convert_robj(pr)
s = pd.Series(df)
aux = s.values
dim = aux.shape[0]
y_pred = np.zeros((len(y_test),dim))

dim = aux.shape[0]
y_pred = np.zeros((len(y_test),dim))
for i in range(dim):
    y_pred[:,i] = np.copy(aux[i]).reshape(1,-1)
y_pred_flx = np.zeros(len(y_test))
for i in range (len(y_pred_flx)):
    y_pred_flx[i] = y_pred[i, flexmix_pred_lbls[i]-1]
flx_mse = calc_error(y_test, y_pred_flx)    
```

```python
print('------------------------------------------')
print 'RR mse:        ', rr_mse
print 'SVR mse:       ', svr_mse
print 'KMeans mse:    ', krr_mse
print 'FlexMix mse:   ', flx_mse
print 'LRR  mse:      ', lrr_mse
print('------------------------------------------')
```

```python
# Create lrr, kmeans, and flexmix matrices with facies results
lrr_mat = np.zeros((facies.sizeX,facies.sizeY))
kmeans_mat = np.zeros((facies.sizeX,facies.sizeY))
flexmix_mat = np.zeros((facies.sizeX,facies.sizeY))
n = len(locXY_train)
for k in range(n):
    x = locXY_train[k][0]
    y = locXY_train[k][1]
    lrr_mat[x,y] = lrr_fit_lbls[k]
    kmeans_mat[x,y] = kmeans_fit_lbls[k]
    flexmix_mat[x,y] = flexmix_fit_lbls[k]-1

n = len(locXY_test)
for k in range(n):
    x = locXY_test[k][0]
    y = locXY_test[k][1]
    lrr_mat[x,y] = lrr_pred_lbls[k]
    kmeans_mat[x,y] = kmeans_pred_lbls[k]
    flexmix_mat[x,y] = flexmix_pred_lbls[k]-1
```

```python
print('------------------------------------------')
print 'LRR predicted facies:       ', np.unique(lrr_mat).astype(int)
print 'KMeans+RR predicted facies: ', np.unique(kmeans_mat).astype(int)
print 'FlexMix predicted facies:   ', np.unique(flexmix_mat).astype(int)
print('------------------------------------------')
```
